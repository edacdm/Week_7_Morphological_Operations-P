```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Kod kullanarak görüntü oluşturma

# Öncelikle içi sıfır ile dolu 300x300 boyutunda 3 kanallı bir
# görüntü matrisi oluşturan kod

width, height, channel = 300, 300, 3

my_red_img = np.zeros((width, height, channel),dtype=np.uint8)

# my_red_img[genişlik,yükseklik,channel]
# RGB: Red, Green, Blue
my_red_img[:,:,0] = 255

plt.imshow(my_red_img)
plt.axis('off')
plt.title("RED IMAGE")
```

```python
my_green_img = np.zeros((width, height, channel),dtype=np.uint8)
my_green_img[:,:,1] = 255

my_blue_img = np.zeros((width, height, channel),dtype=np.uint8)
my_blue_img[:,:,2] = 255

fig = plt.figure(figsize=(10, 7))
rows = 1
columns = 3

fig.add_subplot(rows, columns, 1)
plt.imshow(my_red_img, cmap="gray")
plt.axis('off')
plt.title("RED Image")

fig.add_subplot(rows, columns, 2)
plt.imshow(my_green_img, cmap="gray")
plt.axis('off')
plt.title("GREEN Image")

fig.add_subplot(rows, columns, 3)
plt.imshow(my_blue_img, cmap="gray")
plt.axis('off')
plt.title("BLUE Image")
```

```python
# Beyaz rengini oluşturan kod

my_img_white = np.ones((width, height, channel),dtype=np.uint8)*255
my_img_black = np.zeros((width, height, channel),dtype=np.uint8)

my_img_gray_1 = np.random.randint(128,129,(width, height, channel), dtype=np.uint8)
my_img_gray_2 = np.zeros((width, height, channel),dtype=np.uint8)
my_img_gray_2[:,:,:] = 128

fig = plt.figure(figsize=(10, 7),facecolor='purple')
rows = 2
columns = 2

fig.add_subplot(rows, columns, 1)
plt.imshow(my_img_white, cmap="gray")
plt.axis('off')
plt.title("WHITE Image")

fig.add_subplot(rows, columns, 2)
plt.imshow(my_img_black, cmap="gray")
plt.axis('off')
plt.title("BLACK Image")

fig.add_subplot(rows, columns, 3)
plt.imshow(my_img_gray_1, cmap="gray")
plt.axis('off')
plt.title("GRAY-1 Image")

fig.add_subplot(rows, columns, 4)
plt.imshow(my_img_gray_2, cmap="gray")
plt.axis('off')
plt.title("GRAY-2 Image")
```

```python
# ALMANYA BAYRAĞI ÇİZME

new_width, new_height = 300,450

flag_germany = np.zeros((new_width,new_height,channel),dtype=np.uint8)

# İlk 100 satırı siyah yapan kod
flag_germany[0:100,:] = [0,0,0]

# 100. satırdan 200. satıra kadar kırmızı yapan kod
flag_germany[100:200,:] = [255,0,0]

# 200. satırdan 300. satıra kadar sarı yapan kod
flag_germany[200:300,:] = [255,255,0]

plt.imshow(flag_germany)
plt.axis('off')
```

```python
# BELÇİKA BAYRAĞI ÇİZME (DİKEY BAYRAK)

new_width, new_height = 300,450

flag_belgium = np.zeros((new_width,new_height,channel),dtype=np.uint8)

# İlk 100 sütunu siyah yapan kod
flag_belgium[:,0:150] = [0,0,0]

# 100. sütundan 300. sütuna kadar sarı yapan kod
flag_belgium[:,150:300] = [255,255,0]


# 300. sütundan 450. sütuna kadar kırmızı yapan kod
flag_belgium[:,300:450] = [255,0,0]

plt.imshow(flag_belgium)
plt.axis('off')
```

```python
# MASKELEME (Image Masking), bir görüntüdeki herhangi bir rengin başka bir
# renge dönüştürülmesidir.

# Bunun için öncelikle, değiştirilmek istenen rengin piksel değerleri bir
# değişkene aktarılır. Bu maskedir.

# Daha sonra görüntüde bu renge karşılık gelen koordinatlar belirlenir
# görüntü[maske_koordinatları] = [yeni görüntü renk değerleri]
# yukarıdaki şekilde yeni renk piksellerinin ataması yapılır ve 
# görüntünün istenilen rengi, başka bir renk ile değiştirilmiş olur
```

```python
# BELÇİKA BAYRAĞI ÇİZME (DİKEY BAYRAK)

new_width, new_height = 300,450

flag_belgium = np.zeros((new_width,new_height,channel),dtype=np.uint8)

# İlk 100 sütunu siyah yapan kod
flag_belgium[:,0:150] = [0,0,0]

# 100. sütundan 300. sütuna kadar sarı yapan kod
flag_belgium[:,150:300] = [255,255,0]


# 300. sütundan 450. sütuna kadar kırmızı yapan kod
flag_belgium[:,300:450] = [255,0,0]


# Belçika bayrağındaki sarı rengin mavi olarak değiştirilmesi:
# Öncelikle belçika bayrağındaki sarı renkleri veren pikselleri seçiyoruz

mask_yellow = (flag_belgium[:,:,0]==255)&(flag_belgium[:,:,1]==255)&
(flag_belgium[:,:,2]==0)

# Sarı renklerin olduğu bölgenin mavi renk ile yer değiştirmesi
flag_belgium[mask_yellow] = [0,0,255]

plt.imshow(flag_belgium)
plt.axis('off')
```
